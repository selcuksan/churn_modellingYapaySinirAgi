## BAĞIMLI VE BAĞIMSIZ DEĞİŞKENLERİN ATANMASI

```
df  = df.drop(["RowNumber", "CustomerId", "Surname"],axis=1)
y  = df["Exited"]
x  = df.drop(["Exited"],axis=1,inplace=True)
gender = df["Gender"].values
geography = df[["Geography"]].values
```

```
gender
```

## CİNSİYETLERİN ENCODE EDİLMESİ 0-1

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
gender = le.fit_transform(gender) 
```

```
gender
```

```
gender_Df = pd.DataFrame(gender,columns=["Gender"])
gender_Df
```

## ÜLKELERİN ENCODE EDİLMESİ

```
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
```

```
df["Geography"].unique()
```

```
geography
```

```
ulkeler = ohe.fit_transform(geography).toarray()
```

```
ulkeler
```

```
ulkeler_Df = pd.DataFrame(ulkeler,columns=["France","Germany","Spain"])
ulkeler_Df
```

```
x  = df.drop(["Gender","Geography"],axis=1)
x = pd.concat([ulkeler_Df,gender_Df,x],axis=1)
```

```
x
```

## VERİLERİN EĞİTİM VE TEST OLARAK AYRILMASI

```
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

## BAĞIMSIZ DEĞİŞKENLERİN ÖLÇEKLENDİRİLMESİ

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

## GEREKLİ KÜTÜPHANELERİN İMPORT EDİLMESİ

```
import keras
from keras.models import Sequential
from keras.layers import Dense
```

## YSA VE KATMANLARIN OLUŞTURULMASI

```
model = Sequential()
model.add(Dense(6, kernel_initializer= "uniform", activation="relu", input_dim = 12))
```

```
model.add(Dense(6, kernel_initializer= "uniform", activation="relu"))
```

```
model.add(Dense(1, kernel_initializer= "uniform", activation="sigmoid"))
```

## MODELİN DERLENMESİ

```
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

## MODELİN EĞİTİLMESİ

```
model.fit(x_train,y_train, epochs=50)
```

## TAHMİNLERİN YAPILMASI

```
tahmin = model.predict(x_test)
tahmin = tahmin > 0.5
tahmin
```

## TAHMİN SONUÇLARI

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,tahmin)
cm
```

```
test_loss, test_accuracy = model.evaluate(x_test,y_test)
```
